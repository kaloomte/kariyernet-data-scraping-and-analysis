```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sb
from pylab import rcParams
```

```python
#set some display modifications 
%matplotlib inline
rcParams['figure.figsize'] = 8,4
plt.style.use('seaborn-whitegrid')
pd.set_option('display.max_rows', 1597)
```

```python
#load dataset
df = pd.read_csv('general_dataset.csv')
df.head(15)
```

```python
#remove dot in salary values
df['mean_salary'] = df['mean_salary'].apply(lambda x: x.replace('.',''))
df['max_salary'] = df['max_salary'].apply(lambda x: x.replace('.',''))
df['min_salary'] = df['min_salary'].apply(lambda x: x.replace('.',''))
df['total_participants'] = df['total_participants'].astype(str).apply(lambda x: x.replace('.',''))
#replace '-' with NaN
df = df.apply(lambda x: x.replace('-',np.nan))
df = df.apply(lambda x: x.replace(' -',np.nan))
df = df.apply(lambda x: x.replace('- ',np.nan))
#in 'job_adv.' column there are some values which is very high (37205 this is sum of all advertisement)
#changed with '0'
df['job_adv.'] = df['job_adv.'].astype(str).apply(lambda x: x.split('.')[0].replace('37','0'))
```

```python
#some skills expand the 'number_of_entries ' column
#change them with '0'
df['number_of_entries '][62, 761, 952, 272] = 0
```

```python
df = df.fillna(0)
```

```python
#some skills expand the 'number_of_entries ' column
#change them with '0'
df['number_of_entries '][62, 761, 952, 272] = 0
```

```python
#'male' and 'female' columns has string and percantage types data
#changed into float and rounded
df['male'] = df['male'].astype(float).apply(lambda x: round(x))
df['female'] = df['female'].astype(float).apply(lambda x: round(x))
```

```python
#changed some columns which should be numeric but in string format
df[["mean_salary", "max_salary", 'min_salary', 'number_of_entries ', 'job_adv.', 'total_participants']] = df[["mean_salary", "max_salary", 'min_salary', 'number_of_entries ', 'job_adv.', 'total_participants' ]].apply(pd.to_numeric)
```

```python
df.info()
```

```python
#there are some spaces begin and end of values
#they have been remove
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#prettify some columns names
df = df.rename(columns={'number_of_entries': 'number_of_entries', 'job_adv.': 'job_advertisement','male':'male_percentage','female':'female_percentage'})
```

```python
#sent to clean dataset as a csv file for visualization
df.to_csv('cleandataset.csv', index=False, encoding='utf-8-sig')
```
